# Distribution Plot

```python
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data mining & ML/ford.csv')
```

```python
df.head()
```

```python
sns.distplot(df['price'])
```

```python
sns.distplot(df['price'],kde=False,bins=30)
```

```python
sns.jointplot(x='price',y='mileage',data=df,kind='scatter')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='transmission',palette='coolwarm')
```

# Categorical Data Plots

```python
sns.barplot(x='transmission',y='price',data=df)
```

```python
sns.countplot(x='fuelType',data=df)
```

```python
sns.countplot(x='fuelType',data=df,hue="transmission")
```
